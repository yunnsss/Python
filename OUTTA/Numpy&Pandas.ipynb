## **1. 배열 생성 및 배열 형태 변환**

### 유형1) Python의 list 혹은 tuple을 ndarray로 바꾸기

````python
# [1, 2, 3, 4, 5, 6] list를 ndarray로 바꾸기
list1 = [1, 2, 3, 4, 5, 6]
ndarray1 = np.array(list1)
print(ndarray1)
````

## **4. 인덱싱과 슬라이싱**

- 인덱스: 여러 원소를 가지고 있는 객체(리스트, 튜플, 배열 등)에 부여되는 번호
- 슬라이싱: 여러 원소를 가지고 있는 객체의 일부를 추출하는 것


---


- 파이썬의 List와 동일하게 ndarray도 인덱싱과 슬라이싱 사용 가능

### 1) Indexing (인덱싱)

````python
# 1부터 10까지 수 중 랜덤하게 6개의 수가 추출된 'arr1'이라는 배열
# shape: 3X2인 2차원 배열
arr1 = np.random.randint(1, 11, size = (3, 2))
print(arr1)
````

````python
# 1행 출력
print(arr1[0, :])
````

````python
# 2행 1열의 원소 출력
print(arr1[1, 0])
````

### 2) Slicing (슬라이싱)

- 1차원 배열의 슬라이싱

````python
# 10부터 19까지 연속적인 수로 이루어진 'arr2'라는 배열
# shape: 1차원 배열
arr2 = np.arange(10, 20)
print(arr2)
````

````python
# index가 4부터 6까지인 원소 출력
print(arr2[4:7])
````

````python
# index가 6 이상인 원소 출력
print(arr2[6:])
````

````python
# index가 6 이상인 원소를 모두 20으로 바꾸기
arr2[6:] = 20
print(arr2)
````

- 다차원 배열의 슬라이싱

````python
# 1부터 9까지 연속적인 수로 이루어진 'arr3'라는 배열
# shape: 3X3인 2차원 배열
arr3 = np.arange(1, 10).reshape(3, 3)
print(arr3)
````

````python
# 2, 3, 5, 6으로 구성된 2차원 배열 출력
print(arr3[:2, 1:])
````

````python
# 1행과 2행 출력
print(arr3[0])
print(arr3[0, :])

print(arr3[1])
print(arr3[1, :])
````

````python
# 2열 출력
print(arr3[:, 1])
````

````python
# 2, 3, 5, 6을 모두 결측치로 바꾸기
# 결측치: 데이터에 값이 없는 것 (NaN, null, None)
arr3[:2, 1:] = np.nan
print(arr3)
````

````python
# numpy에서 자료형이 float일 때만 결측치 삽입 가능
arr3 = arr3.astype(float)
arr3[:2, 1:] = np.nan
print(arr3)
````

### 3) 마스킹 연산 (boolean 인덱싱)

- 마스킹 연산: 조건에 부합하는 데이터만 추출하는 방법

````python
# 난수가 표준정규분포 따르는 'arr4'라는 배열 생성
# shape: 5X5인 2차원 배열
arr4 = np.random.randn(5, 5)
print(arr4)
````

````python
# arr4 안의 원소가 양수이면 True, 음수이면 False 출력
print(arr4 > 0)
````

````python
# arr4 안의 원소가 음수면 0 삽입
arr4 = np.where(arr4 > 0, arr4, 0) # 0은 broadcasting 되었다.
print(arr4)
````

### 4) Series 데이터 연산

- 단일 Series내 연산

````python
# 피카츄의 HP가 2배 증가
# 데이터 선택 방법: loc을 사용한 인덱싱
obj['HP'] = obj.loc['HP'] * 2
print(obj)
````

````python
# HP, 공격, 방어 총합 계산
# 데이터 선택 방법: loc을 사용한 슬라이싱
# 집계함수 활용
print(obj.loc['HP':'방어'].sum())
````

````python
# HP(2), 공격, 방어(4) 평균 계산
# 데이터 선택 방법: iloc을 사용한 슬라이싱
# 집계함수 활용
print(obj.iloc[2:5].mean())
````

````python
# 목표: HP가 50미만인 포켓몬들에게 HP+20, Attack+20, Defense+20 하기
# HP가 50 미만인 포켓몬들의 HP, Attack, Defense를 각각 20씩 증가시키는 함수
def increase_stats(row):
    if row['HP'] < 50:
        row['HP'] += 20
        row['Attack'] += 20
        row['Defense'] += 20
    return row

# apply를 사용하여 함수 적용
df.apply(increase_stats, axis=1)
````
