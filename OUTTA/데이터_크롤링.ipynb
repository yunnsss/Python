## í•©ì³ì„œ í¬ë¡¤ë§ í•˜ê¸°

```python
#selenium driver ë¡œë“œ
driver = webdriver.Chrome(options=chrome_options)

#ìƒí’ˆ ì •ë³´ë¥¼ ë°›ì„ ë¦¬ìŠ¤íŠ¸
prod_info_list = []

#ë§í¬ ì „ë‹¬
driver.get("https://www.gmarket.co.kr/n/best")

#CSS_SELECTORë¥¼ í™œìš©í•´ì„œ li íƒœê·¸ë¥¼ ëª¨ë‘ ê°€ì ¸ì˜¤ê¸°
lis = driver.find_elements(By.CSS_SELECTOR, "li")

#lië“¤ì„ ëŒë©´ì„œ
for li in lis:

  try: #itemname ê°€ì ¸ì˜¤ê¸° ì‹œë„í•˜ëŠ”ë°
    itemname = li.find_element(By.CSS_SELECTOR, "a.itemname")
  except: #ì‹¤íŒ¨í•  ê²½ìš°
    continue #ë‹¤ìŒ ë£¨í”„ë¡œ ë„˜ì–´ê°
    #pass #ì­‰ ì§„í–‰

  #sale_price ê°€ì ¸ì˜¤ê¸°
  sale_price = li.find_element(By.CSS_SELECTOR, "div.s-price")
  try:
    original_price = li.find_element(By.CSS_SELECTOR, "div.o-price")
  except:
    original_price = ""
  #prod_info_listì— ìŒ“ê¸°
  prod_info_list.append((itemname.text, sale_price.text, original_price.text))

#driver ì¢…ë£Œ
driver.quit()

print(prod_info_list)
print(len(prod_info_list))
```

```python
gmarket_df = pd.DataFrame(prod_info_list, columns = ['ìƒí’ˆëª…', 'íŒë§¤ê°€', 'ì •ê°€'])

gmarket_df
```

## ì •ê·œí‘œí˜„ì‹ìœ¼ë¡œ ì „ì²˜ë¦¬í•˜ê¸°

```python
import re

re.findall("\d.*ì›", "í• ì¸ê°€ 4,920ì› 2%")
```

```python
def extract_number(x):
    price = re.findall("\d.*ì›", x) #ìˆ«ìë¡œ ì‹œì‘í•´ì„œ ì›ìœ¼ë¡œ ëë‚˜ëŠ” íŒ¨í„´ì„ ì°¾ê¸°
    #ë°ì´í„°ê°€ ë¹ˆ ê²½ìš°ê°€ ìˆìœ¼ë¯€ë¡œ try, exceptë¬¸ ì‚¬ìš©
    try:
      price_num = price[0].replace("ì›", "") #replaceë¡œ ì›ì„ ì œê±°
      #ì²œì› ì•„ë˜ ê°€ê²©ë„ ìˆì„ ìˆ˜ ìˆìœ¼ë¯€ë¡œ try, exceptë¬¸ ì‚¬ìš©
      try:
        price_num = int(price_num.replace(",", "")) #replaceë¡œ ì‰¼í‘œê¹Œì§€ ì œê±°í•˜ê³  intë¡œ ë³€í™˜
      except:
        price_num = int(price_num) #ì•„ë‹ ê²½ìš° ê·¸ëƒ¥ intë¡œ ë³€í™˜
    except:
      pass
    return price_num if price else None #ë°ì´í„°ê°€ ìˆì„ ê²½ìš° price_numì„ ë°˜í™˜, ì•„ë‹ ê²½ìš° Noneì„ ë°˜í™˜
```

```python
gmarket_df['ì •ê°€'] = gmarket_df['ì •ê°€'].apply(extract_number)
gmarket_df['íŒë§¤ê°€'] = gmarket_df['íŒë§¤ê°€'].apply(extract_number)
```

```python
gmarket_df
```

```python
#ê° í–‰ì˜ ì •ê°€ ì—´ì´ naë©´ íŒë§¤ê°€ë¥¼ ëŒ€ì…
def fill_na_with_sales(row):
    if pd.isna(row['ì •ê°€']):
        return row['íŒë§¤ê°€']
    else:
        return row['ì •ê°€']
```

```python
gmarket_df['ì •ê°€'] = gmarket_df.apply(fill_na_with_sales, axis=1)
gmarket_df['ì •ê°€'] = gmarket_df['ì •ê°€'].apply(int)
```

```python
#selenium driver ë¡œë“œ
driver = webdriver.Chrome(options=chrome_options)

#ìƒí’ˆ ì •ë³´ë¥¼ ë°›ì„ ë¦¬ìŠ¤íŠ¸
prod_info_list = []

#ë§í¬ ì „ë‹¬
driver.get("https://www.gmarket.co.kr/n/best?viewType=G&groupCode=G01")

#CSS_SELECTORë¥¼ í™œìš©í•´ì„œ li íƒœê·¸ë¥¼ ëª¨ë‘ ê°€ì ¸ì˜¤ê¸°
lis = driver.find_elements(By.CSS_SELECTOR, "li")

#lië“¤ì„ ëŒë©´ì„œ
for li in lis:

  try: #itemname ê°€ì ¸ì˜¤ê¸° ì‹œë„í•˜ëŠ”ë°
    itemname = li.find_element(By.CSS_SELECTOR, "a.itemname")
  except: #ì‹¤íŒ¨í•  ê²½ìš°
    continue #ë‹¤ìŒ ë£¨í”„ë¡œ ë„˜ì–´ê°
    #pass #ì­‰ ì§„í–‰

  #sale_price ê°€ì ¸ì˜¤ê¸°
  sale_price = li.find_element(By.CSS_SELECTOR, "div.s-price")
  try:
    original_price = li.find_element(By.CSS_SELECTOR, "div.o-price")
  except:
    original_price = ""
  #prod_info_listì— ìŒ“ê¸°
  prod_info_list.append((itemname.text, sale_price.text, original_price.text))

#driver ì¢…ë£Œ
driver.quit()

print(prod_info_list)
print(len(prod_info_list))
```

```python
import re

re.findall("\d.*ì›", "í• ì¸ê°€ 4,920ì› 2%")
```

```python
def extract_number(x):
    price = re.findall("\d.*ì›", x) #ìˆ«ìë¡œ ì‹œì‘í•´ì„œ ì›ìœ¼ë¡œ ëë‚˜ëŠ” íŒ¨í„´ì„ ì°¾ê¸°
    #ë°ì´í„°ê°€ ë¹ˆ ê²½ìš°ê°€ ìˆìœ¼ë¯€ë¡œ try, exceptë¬¸ ì‚¬ìš©
    try:
      price_num = price[0].replace("ì›", "") #replaceë¡œ ì›ì„ ì œê±°
      #ì²œì› ì•„ë˜ ê°€ê²©ë„ ìˆì„ ìˆ˜ ìˆìœ¼ë¯€ë¡œ try, exceptë¬¸ ì‚¬ìš©
      try:
        price_num = int(price_num.replace(",", "")) #replaceë¡œ ì‰¼í‘œê¹Œì§€ ì œê±°í•˜ê³  intë¡œ ë³€í™˜
      except:
        price_num = int(price_num) #ì•„ë‹ ê²½ìš° ê·¸ëƒ¥ intë¡œ ë³€í™˜
    except:
      pass
    return price_num if price else None #ë°ì´í„°ê°€ ìˆì„ ê²½ìš° price_numì„ ë°˜í™˜, ì•„ë‹ ê²½ìš° Noneì„ ë°˜í™˜
```

```python
gmarket_df['ì •ê°€'] = gmarket_df['ì •ê°€'].apply(extract_number)
gmarket_df['íŒë§¤ê°€'] = gmarket_df['íŒë§¤ê°€'].apply(extract_number)
```

```python
#ê° í–‰ì˜ ì •ê°€ ì—´ì´ naë©´ íŒë§¤ê°€ë¥¼ ëŒ€ì…
def fill_na_with_sales(row):
    if pd.isna(row['ì •ê°€']):
        return row['íŒë§¤ê°€']
    else:
        return row['ì •ê°€']
```
